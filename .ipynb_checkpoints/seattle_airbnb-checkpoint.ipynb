# Data Preparation

```python
#columns present in the listings data set
df_listings.columns
```

```python
#Selecting columns
#Columns with numeric or categorical values are selected and those with comments or descriptions are not selected.
df_listings_selected = df_listings[['id','host_id','host_since', 'host_is_superhost','property_type', 'room_type', 'accommodates',
       'bathrooms', 'bedrooms', 'beds', 'bed_type',
       'price', 'weekly_price', 'monthly_price', 'security_deposit',
       'cleaning_fee', 'guests_included', 'extra_people', 'minimum_nights',
       'maximum_nights','availability_30', 'availability_60', 'availability_90',
       'availability_365', 'calendar_last_scraped', 'number_of_reviews',
       'first_review', 'last_review', 'review_scores_rating',
       'review_scores_accuracy', 'review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication',
       'review_scores_location', 'review_scores_value',"cancellation_policy" ]]
```

```python
#Grouping by listing id and counting the summing up all for which the available is t gives the occupied and considering both t and f gives the available available gives the 
df_occupied = pd.DataFrame(pd.pivot_table(df_calendar.groupby(['listing_id', 'available']).count()['date'].reset_index(),index=["listing_id"], columns='available', values='date').reset_index(), columns=['listing_id', 'f', 't']).fillna(0)
df_occupied.columns = ['listing_id', 'occupied', 'available']

df_occupied['occupancy_rate'] = df_occupied['occupied'] / (df_occupied['available'] + df_occupied['occupied'])

#Calculating the % of the occupied
df_occupied.drop(['available', 'occupied'], axis=1, inplace=True)
```

```python
#merging both the listing data set and df_occupied so that we gave occupied percentage in the same row.
data = pd.merge(df_listings_selected, df_occupied, left_on='id', right_on='listing_id',how = 'inner')
data.drop(["listing_id"],axis=1,inplace=True)
```

```python
#Price columns have $ at the beging which needs to be removed.
cols_to_check = ['price', 'weekly_price', 'monthly_price', 'security_deposit', 'cleaning_fee', 'extra_people']
for col in cols_to_check:
    data[col] = data[col].apply(lambda x : x[1:].replace(",", "") if str(x).startswith("$") else x)

data[cols_to_check] = data[cols_to_check].astype(float)
```

```python
# Some rows have missing weekly and monthly price.
# Missing weekly_price and monthly_price are replaced by 7 times the price and 4.34524 times the weekly price respectively.
data['weekly_price'] = data.apply(
    lambda row: row['price']*7 if np.isnan(row['weekly_price']) else row['weekly_price'],
    axis=1
)


data['monthly_price'] = data.apply(
    lambda row: row['weekly_price']*4.34524 if np.isnan(row['monthly_price']) else row['monthly_price'],
    axis=1
)
```

```python
#Columns available now
data.columns
```

```python
data.head()
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#Dropping rows which contains NAs for review_scores_accuracy.
#review_scores_accuracy has been considered because on considering this the NAs with other reviews 
#like review_scores_rating,etc are also removed
data.dropna(subset=["review_scores_accuracy"],inplace=True)                         
```

```python
#Even after thisvariables like security_deposit, cleaning_fee had missing values.
#Assuming that those hosts did not collect cleaning fee and security deposit they have been set to 0
data = data.fillna(0)
```

```python
#Converting strings to dates
data["first_review"] = pd.to_datetime(data["first_review"],format="%Y-%m-%d")
data["last_review"] = pd.to_datetime(data["last_review"],format="%Y-%m-%d")
data["host_since"] = pd.to_datetime(data["host_since"],format="%Y-%m-%d")
df_calendar["date"] = pd.to_datetime(df_calendar["date"],format="%Y-%m-%d")
```

```python
# As the Scrapping of the data was done on 4th Jan 2016, Today is taken as 2016-01-04
Today = pd.to_datetime("2016-01-04",format="%Y-%m-%d")
data["Today"] = [Today]*len(data)
```

```python
#The number of days which has lappsed since the last review that has been given is calculated.
#This column has been created to check if the hosts with less occupation rate also has large today_lastreview value.

data["today_lastreview"] = (data["Today"] - data["last_review"]).dt.days

#The number of days between first review and the last review has been calculated.
#This column has been created to check if the hosts with less occupation rate also has less difference between last and first review dates

data["lastreview_first_review"] = (data["last_review"] - data["first_review"]).dt.days
```

```python
#Function for scatter plot
def scatter_plot(df, X, Y, plt, title=None):
    
    """This is a function which returns scatter plot by taking in the data set, X and Y column names and title
    
    Inputs:
    
    df = data set from which the X and Y axis values are present.
    X = Column name in the data set df which should be the X axis.
    Y = Column name in the data set df which should be the Y axis.
    title = If any title to be specified.
    
    Outputs:
    
    returns Scatter plot
    
    """

    plt.scatter(x=df[X], y=df[Y])

    #If title is None the assign the title as X column name vs Y column name.
    if title is None:
        title = X + ' vs ' + Y
    plt.xlabel(X)
    plt.ylabel(Y)
    plt.title(title)
    
    return plt

#Function for bar plot

def bar_plot(df, X, Y, plt, title=None):
    
    """This is a function which returns bar plot by taking in the data set, X and Y column names and title
    
    Inputs:
    
    df = data set from which the X and Y axis values are present.
    X = Column name in the data set df which should be the X axis.
    Y = Column name in the data set df which should be the Y axis.
    title = If any title to be specified.
    
    Outputs:
    
    returns bar plot
    
    """

    plt.bar(df[X], df[Y])

    
    if title is None:
        title = X + ' vs ' + Y
    plt.xlabel(X)
    plt.ylabel(Y)
    plt.title(title)
    
    return plt
```

```python
# Preparing data to see the behaviour of the occupancy rate with the time.
temp = df_calendar[df_calendar['available'] == 'f'].groupby('date').count()['listing_id'].reset_index()
temp['listing_id'] = temp['listing_id'] / temp.listing_id.nunique()
temp.columns = ['date', 'occupancy_rate']
temp['date'] = pd.to_datetime(temp['date'])
```

```python
#Function to plot the dates.

def plot_date(df, X, Y, plt, title=None):
    
    """This is a function which returns Time series plot by taking in the data set, X and Y column names and title
    
    Inputs:
    
    df = data set from which the X and Y axis values are present.
    X = Column name in the data set df which should be the X axis.
    Y = Column name in the data set df which should be the Y axis.
    title = If any title to be specified.
    
    Outputs:
    
    returns Time series plot
    
    """

    plt.plot_date(x=df[X], y=df[Y], linestyle='solid', marker='None')

    
    if title is None:
        title = X + ' vs ' + Y
    plt.xlabel(X)
    plt.ylabel(Y)
    plt.title(title)
    
    return plt
```

```python
plot_date(temp, 'date','occupancy_rate', plt);
```

This graph shows the occupancy rate with the time.

During Jan there is high occupancy which might be because of holidays. Occupancy decreases as the holidays ends.

There are 2 other peaks, one in April and another in july. This can be because of the spring holidays.

# Amenities importance with occupancy_percentage

```python
#Creating each column for each amenities and assiging it with True if the host provides that amenity. If not, it takes False.

amenities = pd.concat([df_listings.id, df_listings['amenities'].str.get_dummies(sep=',').astype(bool)], axis=1)
```

```python
#Merging the amenities with the id and occupancy rate.

amenities_occupancy_rate = pd.merge(amenities, data[['id','occupancy_rate']], left_on='id', right_on='id',how = 'inner')
```

```python
amenities_occupancy_rate.head()
```

```python
# If All the hosts provide any amenity then removing that amenity.

s = amenities_occupancy_rate.nunique()
unique_data = pd.DataFrame({'column':s.index, 'unique_values':s.values})
```

```python
for idx, row in unique_data.iterrows():
    if row.unique_values == 1:
        amenities_occupancy_rate.drop([row.column], axis=1, inplace=True)
```

```python
#Ploting the corelation matrix.

df = amenities_occupancy_rate

# copy columns to double size of dataframe
df2 = df.copy()
df2.columns = "c_" + df2.columns
df3 = pd.concat([df, df2], axis=1)

# get the correlation coefficient between the different columns
corr = df3.iloc[:, 1:].corr()
arr_corr = corr.as_matrix()
# mask out the top triangle
arr_corr[np.triu_indices_from(arr_corr)] = np.nan

fig, ax = plt.subplots(figsize=(24, 18))

hm = sns.heatmap(arr_corr, cbar=True, vmin=-0.5, vmax=0.5,
                 fmt='.2f', annot_kws={'size': 3}, annot=True, 
                 square=True, cmap=plt.cm.Blues)

ticks = np.arange(corr.shape[0]) + 0.5
ax.set_xticks(ticks)
ax.set_xticklabels(corr.columns, rotation=90, fontsize=8)
ax.set_yticks(ticks)
ax.set_yticklabels(corr.index, rotation=360, fontsize=8)

ax.set_title('correlation matrix')
plt.tight_layout()
plt.savefig("corr_matrix_incl_anno_double.png", dpi=300)
```

This was done to check if proving more/ any amenities attract guests. but there is hardly any correlation between each amenity and the occupancy rate.
